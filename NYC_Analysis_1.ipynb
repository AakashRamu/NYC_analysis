```python
import geopandas as gpd
import pandas as pd
from shapely.wkt import loads
import folium 
```

```python
school=r"C:\Users\aramu\aramu1\Geopandas\geopandas-coures-files\data\New folder\SchoolPoints_APS_2024_08_28.shp"
daycare=r"C:\Users\aramu\aramu1\Geopandas\geopandas-coures-files\data\New folder\DAY_CARE_CENTER_20250209.csv"
parks=r"C:\Users\aramu\aramu1\Geopandas\geopandas-coures-files\data\nyc\Parks Properties_20241223.geojson"
```

```python
school_gdf=gpd.read_file(school)
daycare_df=pd.read_csv(daycare)
parks_gdf=gpd.read_file(parks)
daycare_df.head(5)
```

```python
daycare_df["geometry"]=daycare_df["the_geom"].apply(loads)
daycare_gdf=gpd.GeoDataFrame(daycare_df,geometry="geometry",crs="EPSG:4326")
```

```python
school_gdf.explore(color="red")
parks_gdf.explore()
daycare_gdf.explore()

```

```python

# Create a base map centered around NYC
m = folium.Map(location=[40.7128, -74.0060], zoom_start=12)

# Add School Data (Red Color)
school_gdf.explore(m=m, color="red", tooltip="Name", name="Schools")

# Add Parks Data (Green Color)
parks_gdf.explore(m=m, color="green", tooltip="name311", name="Parks")

# Add Daycare Data (Blue Color)
daycare_gdf.explore(m=m, color="blue", tooltip="NAME", name="Daycare Centers")

# Add Layer Control to Toggle Layers
folium.LayerControl().add_to(m)

# Display the map

m.save(r"C:\Users\aramu\aramu1\Geopandas\geopandas-coures-files\data\New folder\m.html")
```
